## 一、数据挖掘
随着大数据时代的来临，数据越来越庞大呈爆炸性增长，但是给人们带来的挑战也越来越多：
* 一方面，很多信息“隐藏”在数据中，并且并不明显；
* 其次，人工去分析这些数据几乎不可能；
* 再次，很多的数据根本就没有进行分析过；
* ......

人们迫切需要一种技术从海量的数据中获取有用的知识。从中获取的知识可以帮助人们：

* 通过顾客数据分析，可以提供更好的客户服务，定制客户化的服务；
* 可以挖掘更有价值的客户，预测公司明年的前景；
* 可以进行定向营销、商店分布等等；
* 可以对欺诈等进行检测；
* 对大量的科学数据（太空数据、卫星的传感数据、人类基因数据、科学仿真数据...）进行分析，而这是靠人工无法完成的；
* .......  
数据挖掘技术正是在大量数据产生的背景下产生的一种从数据中获取有用知识的技术，特别是大数据时代，大数据挖掘是一个非常重要的应用与技术。

### （一）何为数据挖掘？
**<font color = red>数据挖掘(Data Mining)</font>**, 顾名思义，定义很形象，从“数据”中进行“挖掘”，挖掘一些有用的信息，不同的应用对于挖掘的结果是不一样的。关于“数据挖掘”的定义有很多：
* 从数据中抽取隐含的、以前未知及潜在的有用的信息；
* 利用自动或半自动的方法从大量的数据中探索&分析,发现一些有用的模式
* 数据挖掘是数据库中知识发现（knowledge discovery in database ,KDD）不可缺少的一部分，而KDD是将未加工的数据转换成有用信息的整个过程。

### （二）数据挖掘的步骤
数据挖掘利用各种技术与统计方法，将大量的历史数据，进行整理分析，归纳与整合，是从海量数据中“挖掘”隐藏信息，如趋势、特征及相关的一种过程。BI（商业智能）、数据分析、市场运营等都可以利用数据挖掘技术来完成。

数据挖掘过程模型步骤主要包括：
* 定义问题

明确到底也解决什么问题，不同的问题方法和模型是不一样的，比如说：是分类问题、回归问题还是聚类问题？
* 建立数据挖掘库

建立数据挖掘库包括以下几个步骤：数据收集/采集，数据描述，选择，数据质量评估和数据清理，合并与整合，构建元数据，加载数据挖掘库，维护数据挖掘库。

* 分析数据

分析数据目的是找到对数据挖掘输出影响最大的数据字段，和决定是否需要定义导出字段。数据字段非常多，因此分析数据是一项费力费时的工作，选择好的数据字段进行挖掘，是一个数据挖掘任务结果好坏的一项重要的因素。

* 准备数据（数据预处理）

建立模型之前的最后一步数据准备工作。可以把本步骤分为四个部分：选择变量（特征选择），选择记录，创建新变量，转换变量等。

* 建立模型 

建立模型是一个反复的过程。需要仔细考察不同的模型以判断哪个模型对面对的具体问题最有用。常用的方法是：

先用一部分数据建立模型，然后再用剩下的数据来测试和验证这个得到的模型。

* 评价模型和实施 

模型建立好之后，必须评价得到的结果、解释模型的价值。

### （三）数据挖掘的任务
数据挖掘的任务分为**预测性任务**和**描述性任务**。
* 预测任务：利用一些属性（明变量或自变量）的值来预测其它属性（目标变量或因变量）的值，例如：预测天气、预测房价
* 描述任务：发现人类可以解释的模式来描述数据,例如：关联规则分析

重要的数据挖掘任务如下：
* 分类【Classification】 [Predictive]
* 聚类【Clustering】 [Descriptive]
* 关联规则发现【Association Rule Discovery】 [Descriptive]
* 序列模式发现【Sequential Pattern Discovery】 [Descriptive]
* 回归【Regression】 [Predictive]
* 偏差分析【Deviation Detection】 [Predictive]

## 二、机器学习
机器学习可能是目前最热的技术之一，也是人工智能的一种重要的方法。即时一个非专业的人，一般都听说过机器学习、深度学习等相关概念，但何为“机器学习”？顾名思义，我们人类可以进行学习，在人工智能时代，机器也可以进行学习，我想这也是人工智能走向现实的最关键的一步，让机器自己学习，这是多么“疯狂”的事情。

### 一个例子：“Hi，Siri”唤醒Siri，如何做？  
写一段if语言来唤醒？束手无策？——因为输入时一段语音信号？

![Siri](img/siri.png)
这个时候，机器学习隆重登场:  
(1)收集一个巨大的数据集，里面包含大量的语音信号，每个语音信号是否对应唤醒词。  
(2)不是设计一个系统去准确地辨别唤醒词，而是写一个灵活的程序，带有大量的`参数`，通过调节参数来改变程序的行为，这种程序称为`模型`。

学习(learning)：指在训练过程中更新模型的行为（通过调整参数）。  
通常，用数据训练机器学习模型：
![learning](img/learning.png)

1. 初始化一个什么都不能做的模型  
2. 抓一些有标注的数据集    
3. 修改模型使得它在抓取的数据集上能够更准确执行任务  
4. 重复以上步骤2和3，直到模型看起来不错。

#### 2. 无监督学习
数据没有任何的标签，能从数据中找到某种结构吗？————无监督学习能判断数据属于不同的聚集簇，叫聚类算法。  
例如：  
<ul>
    <li>Google新闻，自动对新闻进行分组。</li>
    <li>给我们⼀堆照⽚，能把它们分成⻛景、狗、婴⼉、猫、⼭峰的照⽚吗？类似的，给定⼀堆⽤⼾浏览记录，我们能把他们分成⼏类典型的⽤⼾吗？</li>
    <li>这类问题通常被称为<b>聚类（clustering）</b>。</li>

</ul>

#### 主成分分析与子空间估计
* 我们可以用少量的参数，准确地捕获数据的相关属性吗？
    * 例如：裁缝们也有⼀组参数，可以准确地描述客⼾的⾝材，以及适合的版型。——这类问题称为`子空间估计（subspace estimation）问题`
* 如果决定因素是线性关系的，则称为`主成分分析（principal component analysis）`。


#### 表征学习（representation learning）
<ul>
    <li>在欧⼏⾥德空间（例如，$R_n$ 中的向量空间）中是否存在⼀种符号属性，可以表⽰出（任意构建的）原始对象？</li>
    <li>例如：城市向量表示“罗⻢-意⼤利+ 法国=巴黎”</li>
    <li>表示学习希望能从数据中自动地学到从数据的原始形式到数据的表示之间的映射</li>
</ul>
例如：word2vec

### （四）机器学习与深度学习
**深度学习是机器学习的一个分支。**深度学习就是一种实现机器学习的优秀技术。深度学习本身是神经网络算法的衍生。深度神经网络本身并不是一个全新的概念，可大致理解为包含多个隐含层的神经网络结构。为了提高深层神经网络的训练效果，人们对神经元的连接方法和激活函数等方面做出相应的调整。

*深度学习的思想，来源于人脑的启发，但绝不是人脑的模拟，人脑不需要大量的数据训练支持，我们只要看过一张猫的图就认识了猫，而机器必须经过几百万张猫的图才能“认识”猫。*

*深度学习，作为目前最热的机器学习方法，但并不意味着是机器学习的终点*



## 三、数据挖掘与机器学习

首先我们看看人工智能、大数据、数据挖掘、机器学习、深度学习间的关系：




具体的，看看数据挖掘与机器学习之间的关系：


数据挖掘和机器学习合在一起讲是因为现在好多数据挖掘的工作是通过机器学习提供的算法工具实现的。例如，
* 广告的ctr预估，PB级别的点击日志在通过典型的机器学习流程可以得到一个预估模型，从而提高互联网广告的点击率和回报率；
* 个性化推荐，还是通过机器学习的一些算法分析平台上的各种购买，浏览和收藏日志，得到一个推荐模型，来预测你喜欢的商品。

我们可以把数据挖掘理解为一种类型的工作，或工作中的某种成分，机器学习是帮助完成这个工作的方法。

**统计学、数据库和人工智能共同构造了数据挖掘技术的三大支柱，许多成熟的统计方法构成了数据挖掘的核心内容。**

通俗一点来讲，从应用落地的角度：
* 数据挖掘是偏数据分析师干的活
* 机器学习是偏算法工程师干的活



1. 从数据上来讲

首先数据传统的数据挖掘数据量并不大(当然现在有大数据挖掘)，只有几万几十万这样，特征一般几百几千个维度；而机器学习，数据分分钟以亿计算，特征以万维计算，比如广告的以几万个商品id为特征，自然语言的以单个词语作为向量特征来做训练等，特征都是上万维。

2. 从算法上来讲

两者的算法有部分重合的，数据挖掘项目能用到关联规则挖掘、聚类、分类、回归等，机器学习算法比这个多而复杂，特别神经网络模型的应用等。

3. 从工作的侧重点来看

数据挖掘的项目很多70%是在做数据的清洗，探索和分析，比如欺诈识别，时间序列预测，多维尺度分析，聚类，因子分析，更偏向于业务洞察。机器学习也许同样很多数据清洗的工作，但它的很多像广告ctr预估、文本挖掘、围棋博弈、视觉识别等都偏工程和数学工作思路。数据挖掘会更喜欢业务应用决策辅助之类，很多时候是统计型的思路，调模型的时候喜欢基于业务洞察做小而美的特征组合、区间缩放之类，机器学习很多时候是加大数据量、把神经网络层数增加等。
